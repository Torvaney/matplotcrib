```python
from cribplot import *
import pandas as pd

%matplotlib inline
```

```python
# Generate some data to use
n = 500  # number of data points

x = np.random.rand(n)  # random x co-ordinates
y = 5 * x + np.random.randn(n)  # y as a function of x with some added noise
z = 3 * x + 0.5 * np.random.randn(n)  # z as a function of x with some added noise
cat = [chr(i) for i in np.random.randint(97, 97+8, n)]  # sample text labels

coords = pd.DataFrame({
        'x': x,
        'y': y,
        'z': z,
        'cat': cat
    })
```

```python
(coords
 .pipe(Plot, figsize=(6, 4))
 .aesthetics(x='x', y='y')
 .points(s=30, alpha=0.5,
         edgecolour=None, colour='k')
 .ref_line(slope=10, intercept=-2, linewidth=5)
 .trendline(linewidth=5)
 .title('Scatter'))
```

```python
p = (coords
     .pipe(Plot, figsize=(12, 8))
     .aesthetics(x='x', y='y', by='cat')
     .points(s=20, alpha=0.5, edgecolour='None', colour='k')
     .trendline(linewidth=2)
)
```

```python
(coords
 .pipe(Plot)
 .aesthetics(x='y', y='cat')
 .points(categorical='y',
         s=30, alpha=1,
         edgecolour='k',
         colour='white')
 .aesthetics(x='x', y='cat')
 .points(categorical='y',
         s=30, alpha=1,
         edgecolour='k',
         colour='blue')
 .title('Categorical scatter'))
```
